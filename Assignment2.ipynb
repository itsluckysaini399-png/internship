
6. Create a list:
arr = [10, 20, 30, 40, 50, 60]
Write a function that takes the list as input and returns the list in
reverse order without using the reverse() method.

```python
arr = [10, 20, 30, 40, 50, 60]

def reverse_list(lst):
    return lst[::-1]

print("Reversed list =", reverse_list(arr))
```

7. Create a tuple:
data = (5, 10, 15, 20, 25, 30, 35)
Write a program to:
○ Count how many elements are divisible by 5
○ Find the sum of all elements
○ Find the average of the tuple

```python
data = (5, 10, 15, 20, 25, 30, 35)

count = 0
total = 0

for i in data:
    if i % 5 == 0:
        count += 1
    total += i

average = total / len(data)

print("Elements divisible by 5 =", count)
print("Sum =", total)
print("Average =", average)
```

8. Create a dictionary:
students = {
"Aman": 78,
"Riya": 92,
"Kriti": 88,
"Rahul": 95
}
Write a program to:
○ Find the student with the highest marks
○ Find the student with the lowest marks
○ Print only the students who scored more than 85 marks

```python
students = {
    "Aman": 78,
    "Riya": 92,
    "Kriti": 88,
    "Rahul": 95
}

max_score = max(students.values())
for name, marks in students.items():
    if marks == max_score:
        print("Top scorer =", name, "with marks", marks)
        
min_score = min(students.values())
for name, marks in students.items():
    if marks == min_score:
        print("Lowest scorer =", name, "with marks", marks)
        
for name, marks in students.items():
    if marks >=85:
        print(name, "has scored above 85 with marks", marks)
```

9. Write a function:
count_frequency(arr)
that takes a list as input and prints the frequency of each element.
Example:
arr = [1, 2, 2, 3, 1, 4, 2]
Output:
1 -> 2 times
2 -> 3 times
3 -> 1 time
4 -> 1 time

```python
def count_frequency(arr):
    for i in set(arr):
        print(i, "->", arr.count(i), "times")

arr = [1, 2, 2, 3, 1, 4, 2]
count_frequency(arr)
```

10.Write a function:
find_duplicates(arr)
that takes a list as input and prints all duplicate elements present in
the list.
Example:
arr = [10, 20, 30, 20, 40, 10, 50, 30]
Output:
Duplicate elements are:
10
20
30

```python
def find_duplicates(arr):
    duplicates = []

    for i in arr:
        if arr.count(i) > 1 and i not in duplicates:
            duplicates.append(i)

    print("Duplicate elements are:")
    for i in duplicates:
        print(i)

arr = [10, 20, 30, 20, 40, 10, 50, 30]
find_duplicates(arr)
```
