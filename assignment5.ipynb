1. Create a 1D NumPy array of 20 random integers between 1 and 100. 
Find the minimum, maximum, mean, and standard deviation without 
using Python built-in functions — use only NumPy aggregation 
functions

```python
import numpy as np
arr = np.random.randint(1, 101,size=20)
print("Array:", arr)
print("Min:", arr.min())
print("Max:", arr.max())
print("Mean:", arr.mean())
print("Std Dev:", arr.std())
```

2. Create a 4×4 matrix using NumPy. Perform the following operations: 
• Extract the diagonal elements
• Replace all even numbers with 0
• Find the index of the maximum value using argmax()

```python
mat = np.random.randint(1, 20, size=(4,4))
print("\nMatrix:\n", mat)
print("Diagonal elements:", np.diag(mat))

mat[mat % 2 == 0] = 0
print("Matrix after replacing even numbers:\n", mat)

print("Index of max value:", mat.argmax())
```

5. Create a DataFrame of at least 10 students with columns: Name, Age, 
Marks, City. Perform the following: 
• Display the first 5 rows using head()
• Show basic statistics using describe()
• Filter students who scored more than 75 marks
• Sort by Marks in descending order

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Lucky","Manas","Karan","Mahi","Bhupendra","Anurag","Gorav","Priya","Mohit","Meena"],
    "Age": [20,21,22,20,23,21,22,20,23,21],
    "Marks": [85,67,90,76,55,88,92,73,60,80],
    "City": ["Delhi","Jaipur","Mumbai","Delhi","Kolkata","Jaipur","Chennai","Delhi","Mumbai","Kolkata"]
}
df = pd.DataFrame(data)

print(df.head())
print(df.describe())
print(df[df["Marks"] > 75])
print(df.sort_values("Marks", ascending=False))
```

6. Create a DataFrame with some missing values intentionally placed in the 
Marks and City columns. Then: 
• Identify missing values using isnull()
• Fill missing marks with the column mean using fillna()
• Drop rows where City is missing using dropna()

```python
df2 = df.copy()
df2.loc[2,"Marks"] = np.nan
df2.loc[4,"City"] = np.nan
print("\nMissing values:\n", df2.isnull())

df2["Marks"].fillna(df2["Marks"].mean(), inplace=True)
df2.dropna(subset=["City"], inplace=True)
print("\nAfter filling and dropping:\n", df2)
```

7. Read a CSV file (you may create it manually) containing product sales 
data with columns: Product, Category, Sales, Region. Perform: 
• Group by Category and calculate total Sales using groupby() and agg()
• Find the top 3 products by sales using sort_values() and head()

```python
sales = pd.DataFrame({
    "Product":["P1","P2","P3","P4","P5"],
    "Category":["Electronics","Clothing","Electronics","Food","Clothing"],
    "Sales":[500,300,700,200,400],
    "Region":["North","South","East","West","North"]
})
print(sales.groupby("Category")["Sales"].agg("sum"))
print(sales.sort_values("Sales", ascending=False).head(3))
```

8. Merge two DataFrames: 
• DataFrame 1: StudentID, Name, Department
• DataFrame 2: StudentID, Marks, Grade
• Merge on StudentID and display students who scored Grade 'A'

```python
df1 = pd.DataFrame({
    "StudentID":[1,2,3],
    "Name":["Lucky","Manas","Mahi"],
    "Department":["CS","IT","ECE"]
})
df2 = pd.DataFrame({
    "StudentID":[1,2,3],
    "Marks":[85,67,90],
    "Grade":["A","B","A"]
})
merged = pd.merge(df1, df2, on="StudentID")
print(merged[merged["Grade"]=="A"])
```

9. Using Matplotlib, create the following plots from a student marks 
dataset (create the dataset yourself): 
• A line plot showing marks trend across 5 subjects
• A bar chart comparing marks of 5 students
• Add proper title, xlabel, ylabel, legend, and grid to both plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
subjects = ["Math","Science","English","History","CS"]
marks = [80,75,90,70,85]

plt.plot(subjects, marks, marker='o', label="Marks Trend")
plt.title("Marks Trend Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()

students = ["Lucky","Manas","Karan","Mahi","Bhupendra"]
marks2 = [85,67,90,76,55]

plt.bar(students, marks2, color='skyblue')
plt.title("Marks Comparison of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(axis='y')
plt.show()
```

10) Using Pandas built-in plotting, load or create a dataset with at least 
4 numeric columns. Create: 
• A histogram for marks distribution
• A box plot to visualize spread and outliers
• A pie chart showing category-wise percentage

```python
df_marks = pd.DataFrame({
    "Math":[80,85,90,70,75],
    "Science":[78,82,88,65,80],
    "English":[85,87,92,70,76],
    "CS":[90,95,85,80,88]
})

df_marks["Math"].plot(kind="hist", title="Math Marks Distribution")
plt.show()

df_marks.plot(kind="box", title="Spread of Marks")
plt.show()

category_data = pd.Series([50,30,20], index=["Electronics","Clothing","Food"])
category_data.plot(kind="pie", autopct='%1.1f%%')
plt.title("Category-wise Percentage")
plt.show()
```

11)Using Seaborn, create a dataset of at least 50 records with columns 
Age, Salary, Department, Experience. Plot: 
• A scatter plot of Age vs Salary with hue as Department
• A heatmap showing correlation between numeric columns
• A bar plot showing average Salary per Department

```python
np.random.seed(42)

df = pd.DataFrame({
    "Age": np.random.randint(22, 50, 50),
    "Salary": np.random.randint(30000, 100000, 50),
    "Department": np.random.choice(
        ["HR", "IT", "Sales", "Finance"], 50
    ),
    "Experience": np.random.randint(1, 20, 50)
})
# Scatter plot of Age vs Salary
sns.scatterplot(data=df, x="Age", y="Salary", hue="Department")
plt.title("Age vs Salary")
plt.show()
corr = df[["Age", "Salary", "Experience"]].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
sns.barplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Average Salary per Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.grid(True)

plt.show()
```
